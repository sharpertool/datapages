```python
descendants = s.get_descendants()
for d in descendants:
    title = d.title.partition('::')[2]
    slug_original = d.slug
    base_slug = slugify(title, allow_unicode=True)
    if base_slug:
        d.slug = d._get_autogenerated_slug(base_slug) 
        d.save()
    print(f"{slug_original} => {d.slug}")
```

```python
def recurse(val):
    if val==0:
        return 0
    else:
        return val+recurse(val-1)
```
